## 1. Line Plot

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")
iris

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.show()
```

```python
# Adding Titles
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("iris plot")
plt.show()
```

## Adding Limits

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("iris plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set Styles

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("iris plot")
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show()
```

## Size of Figure

```python
# change figure,
plt.figure(figsize=(8,6))
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("iris plot")

plt.show()
```

## Load Titanic dataset

```python
# Load dataset
titanic = sns.load_dataset("titanic")
titanic
```

```python
#draw a bar plot
sns.barplot(x="sex", y="alone", hue="who", data=titanic)
plt.show()
```

```python
#draw a bar plot
sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female","male"])
plt.show()
```

```python
#draw a bar plot
sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female","male"], color="green")
plt.show()
```

```python
#draw a bar plot
sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female","male"], color="#B1C241", ci=None,
           palette='pastel')
plt.show()
```

```python
import numpy as np
from numpy import mean
import matplotlib.pyplot as plt

#draw a bar plot
sns.barplot(x="class", y="fare", hue="sex", data=titanic, estimator=mean)
plt.show()
```

```python
#horizontal plot

#draw a bar plot 
sns.barplot(x="fare", y="class", hue="sex", estimator=mean, data=titanic, saturation=1)
plt.show()
```

```python
sns.barplot(x="class", y="fare", data=titanic,
           linewidth=3, facecolor=(.9, 1, 1, 1),
           errcolor=".5", edgecolor="0")
```
