## Data Normalization


- Uniform the data
- they have same impact
- One fish in sea or one in jar
- Also for computation reasons

```python
titanic.head()
```

```python
tn4 = titanic[['age', 'fare']]
tn4.head()
```

- The above is really in wide range and we need to normalize and hard to compare 
- Normalization change the values to the range of 0-to-1(now both variable has similar influence on our models)

## Method of Normalization


1. Simple feature scaling 
   - x(new)=x(old)/x(max)
2. Min-Max method
3. Z-score(standard score)-3-to- +3
4. Log transformation

```python
# simple feature scalling
tn4['fare'] = tn4['fare']/tn4['fare'].max()
tn4['age'] = tn4['age']/tn4['age'].max()
tn4.head()
```

```python
# Min - Max method
tn4['fare'] = (tn4['fare']-tn4['fare'].min())/(tn4['fare'].max()-tn4['fare'].min())
tn4.head()
```

```python
# Z-score (standard score)
tn4['fare'] = (tn4['fare']-tn4['fare'].mean()) / tn4['fare'].std()
tn4.head()
```

```python
tn.head()
```

```python
# log transformation
tn['fare'] = np.log(tn['fare'])
tn.head()
```
